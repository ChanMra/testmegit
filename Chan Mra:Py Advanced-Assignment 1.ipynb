### 1. Missing Values:
   - Identify the columns that contain missing values and count how many missing values are in each column.

```python
df.dtypes
```

```python
avg_age = round (df ['Age'].mean (axis = 0), 0)
```

```python
df['Age'] = df['Age'].replace(np.nan, avg_age).astype (int)
avg_age
```

```python
#The avg age is 41 and it will be replaced in the null-columns
```

```python
df ['Salary'] = df['Salary'].astype(float)
avg_salary = round (df ['Salary'].mean (axis = 0), 2)
avg_salary
```

```python
df ['Salary'] = df ['Salary'].replace (np.nan, avg_salary)
```

### 3. Data Transformation:
   - Create a new column called `SalaryBracket` that bins the `Salary` values into categories: "Low", "Medium", and "High." Use appropriate salary ranges for each category.
   - Justify the ranges you selected for each salary bracket.

```python
bins = np.linspace(min(df["Salary"]), max(df["Salary"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df ['SalaryBracket'] = pd.cut (df ['Salary'], bins, labels = group_names, include_lowest = True)
```

```python
df['SalaryBracket'].value_counts ()
```

#### The salary range is divided among the minimum one third of the range, middle one third of the range and the maximum one third of the range.

### 4. Binning Performance Scores:
   - Divide the `Performance` scores into three bins: "Poor", "Average", and "Excellent." Use `pd.cut()` for this task and define the bin ranges clearly.
   - Explain how binning can make performance data easier to interpret.

```python
bins = np.linspace(min(df["Performance"]), max(df["Performance"]), 4)
bins
```

```python
group_names = ['poor', 'Average', 'Excellence']
```

```python
df ['PerformanceScore'] = pd.cut (df ['Performance'], bins, labels = group_names, include_lowest = True)
```

```python
df['PerformanceScore'].value_counts ()
```

#### Binning Performance can simplify continuous data into categorical data which makes easier to interpret.
